```python
from google.colab.patches import cv2_imshow
import cv2
import numpy as np
my_pic=cv2.imread('my_sig.png')
my_pic=cv2.resize(my_pic,(200,200))
cv2_imshow(my_pic)
```

```python
# brighten the image
brightness=8
a=my_pic+brightness
bright_image=np.clip(a,0,255)
cv2_imshow(bright_image)
```

```python
# rotating
height,width=bright_image.shape[:2]
center=(width/2,height/2)
rotation_matrix=cv2.getRotationMatrix2D(center,10,1)
rotated_image=cv2.warpAffine(bright_image,rotation_matrix,(width,height))
cv2_imshow(rotated_image)
```

```python
for i in range(0,360,20):
  from google.colab.patches import cv2_imshow
  import cv2
  import numpy as np
  my_pic=cv2.imread('my_sig.png')
  my_pic=cv2.resize(my_pic,(200,200))
  # brighten the image
  brightness=8
  a=my_pic+brightness
  bright_image=np.clip(a,0,255)
  # rotating
  height,width=bright_image.shape[:2]
  center=(width/2,height/2)
  rotation_matrix=cv2.getRotationMatrix2D(center,i,1)
  rotated_image=cv2.warpAffine(bright_image,rotation_matrix,(width,height))
  cv2_imshow(rotated_image)
```
