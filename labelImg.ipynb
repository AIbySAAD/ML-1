```python
from google.colab.patches import cv2_imshow
import cv2
img=cv2.imread('1.jpeg')
cv2_imshow(img)
```

```python
img.shape
```

```python
point1=(int(0.659153 * img.shape[1]),int(0.279599*img.shape[0]))
point2=(int(0.353825 * img.shape[1]),int(0.420765*img.shape[0]))
```

```python
rect_img=cv2.rectangle(img,point1,point2,color=(255,0,255),thickness=3)
cv2_imshow(rect_img)
```
